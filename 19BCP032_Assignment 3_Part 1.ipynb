# Data Cleaning

```python
remove_categorical_cols = []
for col in data:
    if data[col].dtypes == "object":
        remove_categorical_cols.append(col)
remove_categorical_cols.append('Id')
data = data.drop(columns = remove_categorical_cols)
data
```

```python
data = data.dropna()
data
```

# Predicting Model by keeping all the Features

```python
X = data.drop(['SalePrice'], axis = 1)
Y = data['SalePrice']
```

```python
X
```

```python
Y
```

```python
print(X.shape)
```

```python
print(Y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Splitting the data into training and testing data
LR = LinearRegression()
LR.fit(x_train, y_train)
y_pred = LR.predict(x_test)
```

# Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
errors = mean_squared_error(y_test, y_pred)
print(errors)
```

# Root Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
errors = mean_squared_error(y_test, y_pred, squared=False)
print(errors)
```

# Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
errors = mean_absolute_error(y_test, y_pred)
print(errors)
```

```python
r2_score(y_test, y_pred)
```
